```python
import pandas as pd
import numpy as np
```

```python
def get_data(station_number,file_prefix='hly',base_url='http://cli.fusio.net/cli/climate_data/webdata/',n_blanks=3,out_dir='data'):
    station_url = f'{base_url}{file_prefix}{station_number}.csv'
    # create df_tmp to figure out where data starts in the files
    df_tmp = pd.read_csv(station_url,nrows=50,sep='$')
    n_skiprows = np.min(df_tmp[df_tmp[df_tmp.columns[0]].str[:8]=='date,ind'].index.values)
    # get header info
    df_header = pd.read_csv(station_url,nrows=n_skiprows,sep='$')
    df_header.to_csv(f'{out_dir}/{file_prefix}{station_number}_header.csv',index=False)
    # get data
    df = pd.read_csv(station_url,skiprows=n_skiprows+n_blanks)
    df = df.set_index('date')
    df = df.apply(pd.to_numeric,errors='coerce')
    df.to_parquet(f'{out_dir}/{file_prefix}{station_number}.parquet')
    
def try_get_data(station_number,n_try=3):
    attempt_number = 0
    while attempt_number < n_try:
        try:
            get_data(station_number)
            break
        except Exception as e:
            attempt_number += 1
            print(f'attempt_number={attempt_number}, {e}')
```

```python
# get station details info
df_station_details = pd.read_csv('data/station_details.csv')
print(df_station_details.shape)
df_station_details.head()
```

```python
for station_number in df_station_details['Station Number'].unique():
    print(f'station_number={station_number}')
    try_get_data(station_number)
```
