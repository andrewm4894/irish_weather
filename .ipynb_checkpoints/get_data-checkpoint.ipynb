```python
import pandas as pd
import numpy as np
```

```python
def get_data(station_number,file_prefix='hly',base_url='http://cli.fusio.net/cli/climate_data/webdata/',n_blanks=3,out_dir='data',pq_compression='gzip'):
    station_url = f'{base_url}{file_prefix}{station_number}.csv'
    # create df_tmp to figure out where data starts in the files
    df_tmp = pd.read_csv(station_url,nrows=50,sep='$')
    n_skiprows = np.min(df_tmp[df_tmp[df_tmp.columns[0]].str[:8]=='date,ind'].index.values)
    # get header info
    df_header = pd.read_csv(station_url,nrows=n_skiprows,sep='$')
    df_header.to_csv(f'{out_dir}/{file_prefix}{station_number}_header.csv',index=False)
    # get data
    df = pd.read_csv(station_url,skiprows=n_skiprows+n_blanks)
    df = df.set_index('date')
    df = df.apply(pd.to_numeric,errors='coerce')
    if pq_compression == 'gzip':
        pq_file_extension = 'gz.parquet'
    else:
        pq_file_extension = 'parquet'
    pq_file_name = f'{out_dir}/{file_prefix}{station_number}.{pq_file_extension}'
    df.to_parquet(pq_file_name,compression=pq_compression)
    
def try_get_data(station_number,n_try=5,file_prefix='hly'):
    attempt_number = 0
    while attempt_number < n_try:
        try:
            get_data(station_number,file_prefix=file_prefix)
            break
        except Exception as e:
            attempt_number += 1
            if attempt_number == n_try:
                print(f'attempt_number={attempt_number}, {e}')
```

```python
# get station details info
df_station_details = pd.read_csv('data/station_details.csv')
print(df_station_details.shape)
df_station_details = df_station_details.sort_values(by='Station Number')
df_station_details.head()
```

```python
# loop over station numbers and try get data
for station_number in df_station_details['Station Number'].unique():
    print(f'station_number={station_number}')
    try_get_data(station_number,file_prefix='dly')
```
